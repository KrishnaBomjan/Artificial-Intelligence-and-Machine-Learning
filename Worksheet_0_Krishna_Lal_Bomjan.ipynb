Task - 1:
Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.
• Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.
• Conversion Options:
1. Length:
– Convert meters (m) to feet (ft).
– Convert feet (ft) to meters (m).
2. Weight:
– Convert kilograms (kg) to pounds (lbs).
– Convert pounds (lbs) to kilograms (kg).
3. Volume:
– Convert liters (L) to gallons (gal).
– Convert gallons (gal) to liters (L).

```python
# Task 1: Unit Converter
def convert_length(value, from_unit, to_unit):
    """
    Converts length between meters and feet.

    Parameters:
        value (float): The value to convert
        from_unit (str): The unit to convert from ('m' or 'ft')
        to_unit (str): The unit to convert to ('m' or 'ft')

    Returns:
        float: The converted value
    """
    if from_unit == 'm' and to_unit == 'ft':
        return value * 3.28084
    elif from_unit == 'ft' and to_unit == 'm':
        return value / 3.28084
    else:
        raise ValueError("Invalid length units. Use 'm' or 'ft'.")

def convert_weight(value, from_unit, to_unit):
    """
    Converts weight between kilograms and pounds.

    Parameters:
        value (float): The value to convert
        from_unit (str): The unit to convert from ('kg' or 'lbs')
        to_unit (str): The unit to convert to ('kg' or 'lbs')

    Returns:
        float: The converted value
    """
    if from_unit == 'kg' and to_unit == 'lbs':
        return value * 2.20462
    elif from_unit == 'lbs' and to_unit == 'kg':
        return value / 2.20462
    else:
        raise ValueError("Invalid weight units. Use 'kg' or 'lbs'.")

def convert_volume(value, from_unit, to_unit):
    """
    Converts volume between liters and gallons.

    Parameters:
        value (float): The value to convert
        from_unit (str): The unit to convert from ('L' or 'gal')
        to_unit (str): The unit to convert to ('L' or 'gal')

    Returns:
        float: The converted value
    """
    if from_unit == 'L' and to_unit == 'gal':
        return value * 0.264172
    elif from_unit == 'gal' and to_unit == 'L':
        return value / 0.264172
    else:
        raise ValueError("Invalid volume units. Use 'L' or 'gal'.")

def unit_converter():
    """
    Main function to handle user interaction for unit conversion.
    """
    print("Unit Converter Program")
    print("1. Length (meters <-> feet)")
    print("2. Weight (kilograms <-> pounds)")
    print("3. Volume (liters <-> gallons)")

    try:
        choice = input("Enter your choice (1-3): ")

        if choice == '1':
            print("\nLength Conversion:")
            print("a. Meters to Feet")
            print("b. Feet to Meters")
            sub_choice = input("Enter choice (a/b): ").lower()
            value = float(input("Enter value: "))

            if sub_choice == 'a':
                result = convert_length(value, 'm', 'ft')
                print(f"{value} meters = {result:.4f} feet")
            elif sub_choice == 'b':
                result = convert_length(value, 'ft', 'm')
                print(f"{value} feet = {result:.4f} meters")
            else:
                print("Invalid choice")

        elif choice == '2':
            print("\nWeight Conversion:")
            print("a. Kilograms to Pounds")
            print("b. Pounds to Kilograms")
            sub_choice = input("Enter choice (a/b): ").lower()
            value = float(input("Enter value: "))

            if sub_choice == 'a':
                result = convert_weight(value, 'kg', 'lbs')
                print(f"{value} kg = {result:.4f} lbs")
            elif sub_choice == 'b':
                result = convert_weight(value, 'lbs', 'kg')
                print(f"{value} lbs = {result:.4f} kg")
            else:
                print("Invalid choice")

        elif choice == '3':
            print("\nVolume Conversion:")
            print("a. Liters to Gallons")
            print("b. Gallons to Liters")
            sub_choice = input("Enter choice (a/b): ").lower()
            value = float(input("Enter value: "))

            if sub_choice == 'a':
                result = convert_volume(value, 'L', 'gal')
                print(f"{value} L = {result:.4f} gallons")
            elif sub_choice == 'b':
                result = convert_volume(value, 'gal', 'L')
                print(f"{value} gallons = {result:.4f} L")
            else:
                print("Invalid choice")
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

unit_converter()
```

Create a Python program that performs various mathematical operations on a list of numbers.
• The Program should:

```python
# Task 2: Mathematical Operations on List

def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
        numbers (list): List of numbers

    Returns:
        float: Sum of the numbers
    """
    return sum(numbers)

def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
        numbers (list): List of numbers

    Returns:
        float: Average of the numbers
    """
    if len(numbers) == 0:
        raise ValueError("Cannot calculate average of empty list")
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    """
    Finds the maximum value in a list.

    Parameters:
        numbers (list): List of numbers

    Returns:
        float: Maximum value
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find maximum of empty list")
    return max(numbers)

def find_minimum(numbers):
    """
    Finds the minimum value in a list.

    Parameters:
        numbers (list): List of numbers

    Returns:
        float: Minimum value
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find minimum of empty list")
    return min(numbers)

def math_operations():
    """
    Main function to handle mathematical operations on a list of numbers.
    """
    print("Mathematical Operations on List")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = input("Enter your choice (1-4): ")

        # Get input and convert to list of numbers
        input_str = input("Enter numbers separated by spaces: ")
        if not input_str.strip():
            raise ValueError("Empty input provided")

        numbers = [float(x) for x in input_str.split()]

        if len(numbers) == 0:
            raise ValueError("No numbers provided")

        if choice == '1':
            result = calculate_sum(numbers)
            print(f"Sum: {result}")
        elif choice == '2':
            result = calculate_average(numbers)
            print(f"Average: {result}")
        elif choice == '3':
            result = find_maximum(numbers)
            print(f"Maximum: {result}")
        elif choice == '4':
            result = find_minimum(numbers)
            print(f"Minimum: {result}")
        else:
            print("Invalid choice")

    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

math_operations()
```

```python
# 1. Extract Every Other Element
def extract_every_other(lst):
    """
    Extracts every other element from a list, starting from the first element.

    Parameters:
        lst (list): Input list

    Returns:
        list: Every other element starting from index 0
    """
    return lst[::2]


print("1. Extract Every Other Element:")
print(extract_every_other([1, 2, 3, 4, 5, 6]))  # Output: [1, 3, 5]


# 2. Slice a Sublist
def get_sublist(lst, start, end):
    """
    Returns a sublist from start to end (inclusive).

    Parameters:
        lst (list): Input list
        start (int): Starting index
        end (int): Ending index

    Returns:
        list: Sublist from start to end (inclusive)
    """
    return lst[start:end+1]


print("\n2. Slice a Sublist:")
print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))  # Output: [3, 4, 5]


# 3. Reverse a List Using Slicing
def reverse_list(lst):
    """
    Reverses a list using slicing.

    Parameters:
        lst (list): Input list

    Returns:
        list: Reversed list
    """
    return lst[::-1]

# Test
print("\n3. Reverse a List:")
print(reverse_list([1, 2, 3, 4, 5]))  # Output: [5, 4, 3, 2, 1]


# 4. Remove the First and Last Elements
def remove_first_last(lst):
    """
    Removes the first and last elements of a list.

    Parameters:
        lst (list): Input list

    Returns:
        list: List without first and last elements
    """
    return lst[1:-1]

print("\n4. Remove First and Last:")
print(remove_first_last([1, 2, 3, 4, 5]))  # Output: [2, 3, 4]


# 5. Get the First n Elements
def get_first_n(lst, n):
    """
    Extracts the first n elements from a list.

    Parameters:
        lst (list): Input list
        n (int): Number of elements to extract

    Returns:
        list: First n elements
    """
    return lst[:n]

print("\n5. Get First n Elements:")
print(get_first_n([1, 2, 3, 4, 5], 3))  # Output: [1, 2, 3]


# 6. Extract Elements from the End
def get_last_n(lst, n):
    """
    Extracts the last n elements of a list.

    Parameters:
        lst (list): Input list
        n (int): Number of elements from the end

    Returns:
        list: Last n elements
    """
    return lst[-n:]

print("\n6. Get Last n Elements:")
print(get_last_n([1, 2, 3, 4, 5], 2))  # Output: [4, 5]


# 7. Extract Elements in Reverse Order
def reverse_skip(lst):
    """
    Extracts every second element starting from the second-to-last, moving backward.

    Parameters:
        lst (list): Input list

    Returns:
        list: Elements in reverse order with skip
    """
    return lst[-2::-2]

print("\n7. Reverse Skip:")
print(reverse_skip([1, 2, 3, 4, 5, 6]))  # Output: [5, 3, 1]
```

```python
# 4.3 Exercise on Nested List

# 1. Flatten a Nested List
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
        lst (list): Nested list

    Returns:
        list: Flattened list
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

# Test
print("1. Flatten Nested List:")
print(flatten([[1, 2], [3, 4], [5]]))  # Output: [1, 2, 3, 4, 5]


# 2. Accessing Nested List Elements
def access_nested_element(lst, indices):
    """
    Extracts a specific element from a nested list given its indices.

    Parameters:
        lst (list): Nested list
        indices (list): List of indices

    Returns:
        The element at the specified position
    """
    current = lst
    for index in indices:
        current = current[index]
    return current

# Test
print("\n2. Access Nested Element:")
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(access_nested_element(lst, [1, 2]))  # Output: 6


# 3. Sum of All Elements in a Nested List
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list.

    Parameters:
        lst (list): Nested list

    Returns:
        float: Sum of all elements
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

# Test
print("\n3. Sum Nested List:")
print(sum_nested([[1, 2], [3, [4, 5]], 6]))  # Output: 21


# 4. Remove Specific Element from a Nested List
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
        lst (list): Nested list
        elem: Element to remove

    Returns:
        list: Modified list with element removed
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            new_sublist = remove_element(item, elem)
            if new_sublist:  # Only add non-empty lists
                result.append(new_sublist)
        elif item != elem:
            result.append(item)
    return result

# Test
print("\n4. Remove Element:")
print(remove_element([[1, 2], [3, 2], [4, 5]], 2))  # Output: [[1], [3], [4, 5]]


# 5. Find the Maximum Element in a Nested List
def find_max(lst):
    """
    Finds the maximum element in a nested list.

    Parameters:
        lst (list): Nested list

    Returns:
        The maximum element
    """
    max_val = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)
    return max_val

# Test
print("\n5. Find Maximum:")
print(find_max([[1, 2], [3, [4, 5]], 6]))  # Output: 6


# 6. Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
        lst (list): Nested list
        elem: Element to count

    Returns:
        int: Count of occurrences
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

# Test
print("\n6. Count Occurrences:")
print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))  # Output: 3


# 7. Flatten a List of Lists of Lists (Deep Flatten)
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.

    Parameters:
        lst (list): Deeply nested list

    Returns:
        list: Flattened list
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

# Test
print("\n7. Deep Flatten:")
print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))  # Output: [1, 2, 3, 4, 5, 6, 7, 8]


# 8. Nested List Average
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
        lst (list): Nested list

    Returns:
        float: Average of all elements
    """
    def sum_and_count(lst):
        total = 0
        count = 0
        for item in lst:
            if isinstance(item, list):
                s, c = sum_and_count(item)
                total += s
                count += c
            else:
                total += item
                count += 1
        return total, count

    total, count = sum_and_count(lst)
    return total / count if count > 0 else 0

# Test
print("\n8. Nested List Average:")
print(average_nested([[1, 2], [3, 4], [5, 6]]))  # Output: 3.5
```

```python
# Problem 4: Matrix Operations
print("\n" + "=" * 50)
print("Problem 4: Matrix Operations")
print("=" * 50)

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print(f"A = \n{A}")
print(f"B = \n{B}")

# 1. Prove A.A^-1 = I
A_inv = np.linalg.inv(A)
product = np.dot(A, A_inv)
print("\n1. Prove A · A^(-1) = I:")
print(f"   A^(-1) = \n{A_inv}")
print(f"   A · A^(-1) = \n{product}")
print(f"   Is it close to I? {np.allclose(product, np.eye(2))}")

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. Prove AB ≠ BA:")
print(f"   AB = \n{AB}")
print(f"   BA = \n{BA}")
print(f"   Are they equal? {np.array_equal(AB, BA)}")

# 3. Prove (AB)^T = B^T · A^T
AB_T = AB.T
BT_AT = np.dot(B.T, A.T)
print("\n3. Prove (AB)^T = B^T · A^T:")
print(f"   (AB)^T = \n{AB_T}")
print(f"   B^T · A^T = \n{BT_AT}")
print(f"   Are they equal? {np.allclose(AB_T, BT_AT)}")

# Solve system of linear equations
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

print("\n" + "=" * 50)
print("Solving System of Linear Equations")
print("=" * 50)

# Coefficient matrix A
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])
# Constants matrix B
B_eq = np.array([-1, -3, 9])

print(f"Coefficient matrix A:\n{A_eq}")
print(f"Constants B: {B_eq}")

# Method 1: Using inverse (A·X = B => X = A^(-1)·B)
A_inv_eq = np.linalg.inv(A_eq)
X_inv = np.dot(A_inv_eq, B_eq)
print(f"\nSolution using inverse method:")
print(f"x = {X_inv[0]:.4f}, y = {X_inv[1]:.4f}, z = {X_inv[2]:.4f}")

# Method 2: Using np.linalg.solve (recommended)
X_solve = np.linalg.solve(A_eq, B_eq)
print(f"\nSolution using np.linalg.solve:")
print(f"x = {X_solve[0]:.4f}, y = {X_solve[1]:.4f}, z = {X_solve[2]:.4f}")

# Verification
verification = np.dot(A_eq, X_solve)
print(f"\nVerification (A·X should equal B): {verification}")
print(f"Original B: {B_eq}")
print(f"Match: {np.allclose(verification, B_eq)}")
```
